```python
import re
from pathlib import Path
name = "john"


input_folder = Path(f"books/{name}/source")
output_folder = Path(f"books/{name}")

output_folder.mkdir(parents=True, exist_ok=True)

def capitalize_first(match):
    word = match.group()
    return word[:1].upper() + word[1:].lower()

def clean_text(text):
    text = re.sub(r"\[.*?\]", "", text)

    text = re.sub(r"господ\w*", "kyrios", text)
    
    text = re.sub(r"бог\w*", capitalize_first, text)
    text = re.sub(r"бож\w*", capitalize_first, text)
    
    text = re.sub(r"мессией\w*", "помазаником", text)
    text = re.sub(r"мессия\w*", "помазаник", text)
    text = re.sub(r"мессию\w*", "помазаника", text)

    # remove dots, commas AND extra punctuation
    text = re.sub(r"[.,:;!\?\—\"«»“„]", "", text)

    return text

files = [
    f for f in input_folder.glob("*.html")
    if f.name.lower() != "index.html"
]

print("Found files:", len(files))

for file_path in files:
    content = file_path.read_text(encoding="utf-8")

    cleaned = content.lower()
    cleaned = clean_text(cleaned)

    output_path = output_folder / file_path.name
    output_path.write_text(cleaned, encoding="utf-8")

    print("Processed:", file_path.name)
```
