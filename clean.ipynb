```python
import re
from pathlib import Path

input_folder = Path(r"john/john_source")
output_folder = Path(r"john")

output_folder.mkdir(parents=True, exist_ok=True)

def clean_text(text):
    text = re.sub(r"\[.*?\]", "", text)

    text = re.sub(r"господ\w*", "kyrios", text)
    text = re.sub(r"бог\w*", "Theos", text)
    text = re.sub(r"бож\w*", "Theos", text)

    return text

files = [
    f for f in input_folder.glob("*.html")
    if f.name.lower() != "index.html"
]

print("Found files:", len(files))

for file_path in files:
    content = file_path.read_text(encoding="utf-8")

    # convert all text to lowercase (your current choice)
    cleaned = content.lower()

    cleaned = clean_text(cleaned)

    output_path = output_folder / file_path.name
    output_path.write_text(cleaned, encoding="utf-8")

    print("Processed:", file_path.name)
```
